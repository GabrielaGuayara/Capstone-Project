```python
import numpy as np
import pandas as pd
import requests as re
import seaborn as sns
from tqdm.auto import tqdm
import matplotlib.pyplot as plt


import plotly
import plotly.express as px
import plotly.offline as py
import plotly.graph_objs as go
```

# Retrieve Data

```python
path="/electricity_data.csv"
df=pd.read_csv(path)
df.head(5)
```

```python
df.describe()
```

# Data Clearning

```python
df.info()
```

```python
# Remove the missing values
df = df.dropna()
```

# Finding 1: Sales per Month in 2022 and 2023

```python
df.info()
```

# Finding 2: Date, Time & Crash Numbers

```python
df = df[df['year'] == 2023]
```

```python
groupy_df = df.groupby(["month", "sector_id"])["sales"].sum().reset_index()
fig = px.bar(groupy_df, x="month", y="sales", color="sector_id", text="sales")
fig.update_layout(barmode='stack')
fig.show()
```
